```python
import pandas as pd

df = pd.read_excel("validate.xlsx")

new_frame = pd.read_json("solidity_vulnerabilities_dataset.json")


```

```python
df.head()
```

```python
new_frame.head()
```

```python
new_frame.rename(columns={'code': 'code', 'vulnerabilities': 'contain', 'fixed': 'miss'}, inplace=True)
```

```python
new_frame.head()
```

```python
import json
for i in range(len(new_frame)):
    new_frame.at[i, 'contain'] = json.dumps(new_frame.at[i, 'contain'])
    new_frame.at[i, 'miss'] = json.dumps(new_frame.at[i, 'miss'])
```

```python
new_frame.head()
```

```python
df = pd.concat([df, new_frame], ignore_index=True)
```

```python
df.to_excel("validate.xlsx", index=False)
```
